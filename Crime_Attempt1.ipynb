```python
import pandas as pd
import numpy as np
import matplotlib as plt

#Library to use to read a PDF
import PyPDF2
```

```python
#Opening the PDF to be read
f = open('Resources/Citywide_Reported_Offenses_2020.pdf','rb')
```

```python
#Number of pages to be read in the PDF
pdf_reader = PyPDF2.PdfFileReader(f)
pdf_reader.numPages
```

```python
#Reading Page 1 of the PDF
page_one = pdf_reader.getPage(0)
page_two = pdf_reader.getPage(1)
```

```python
#Extracting the text from page 1
page_one_text = page_one.extractText()
page_one_text
```

```python
print(page_one_text)
```

```python
#Extracting the text from page 2
page_two_text = page_two.extractText()
page_two_text
```

```python
print(page_two_text)
```

```python
#Closes the opened file. A closed file cannot be read or written any more.
f.close()
```
